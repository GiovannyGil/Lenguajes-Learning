```python
import matplotlib.pyplot as plt
```

COMPATIBLE CON PANDAS, NUMPY Y SUS ELEMENTOS COMO ARREGLOS

```python
# ejempĺo de ima linea simple
x = [0,1,2,3,4,5]
y = [0,1,2,3,4,5]

plt.plot(x,y)
plt.scatter(x,y)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
# plt.show()

```

```python
# Grafico de barras

categorias = ["A", "B", "C", "D"]
valores = [10,20,25,30]

plt.bar(categorias, valores)
plt.title("Gráfico de Barras")
plt.xlabel("Categorias")
plt.ylabel("Valores")
plt.show()
```

```python
x = [0,1,2,3,4,5]
y = [0,1,2,9,16,25]

plt.plot(x,y, color="red", linestyle = "--", marker="o")
plt.xlabel("Valores Medidos")
plt.ylabel("Tiempo")
plt.show()
```
